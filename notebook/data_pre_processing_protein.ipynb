```r
# Load necessary libraries.
library(readxl)
library(tibble)
library(dplyr)
library(tidyr)
```

```r
# Helpers functions.
get_line_id <- function(str) {
    return(strsplit(str, split = " ")[[1]][1])
}
```

```r
# Read raw data.
raw_data <- data.frame()
xlsx_file_path <- "../data/raw_data/Supplementary Data 1.xlsx"
for (sheet_name in readxl::excel_sheets(xlsx_file_path)) {
    data_sheet <- readxl::read_excel(xlsx_file_path, sheet = sheet_name)
    data_sheet['line_id'] <- get_line_id(sheet_name)
    raw_data <- rbind(raw_data, data_sheet)
}
```

```r
# QUESTION : We can use smaller cut_off, more data.
commonity_cut_off <- 8
protein_ids_count <- table(raw_data$Accession)
protein_ids_count <- protein_ids_count[protein_ids_count >= commonity_cut_off]
protein_ids <- names(protein_ids_count)
```

```r
common_data <- raw_data[raw_data$Accession %in% protein_ids,]
```

```r
common_data <- na.omit(common_data)
```

```r
common_data_summary <- common_data %>% group_by(Accession, line_id) %>% summarise(score_mean=(mean(Score)))
```

```r
data <- common_data_summary %>% pivot_wider(names_from = line_id, values_from = score_mean) %>% column_to_rownames(var = "Accession")
```

```r
write.csv(data, "../data/clean_data/proteins_score_data.csv", row.names = TRUE)
```
