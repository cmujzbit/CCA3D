```r
# Read raw data.
raw_data <- read.csv("../data/raw_data/summary-data.csv")
```

```r
not_primary_transcript_mask <- raw_data["type"] != "miRNA_primary_transcript"
data_not_primary_transcript <- raw_data[not_primary_transcript_mask, ]
```

```r
raw_data_col_names <- c("sRNA_115", "sRNA_266_4", "sRNA_793", "sRNA_LU")
data_line_names <- c("WM115", "WM266-4", "WM793", "WM1205Lu")
data <- data_not_primary_transcript[raw_data_col_names]
row.names(data) <- data_not_primary_transcript[["ID"]]
colnames(data) <- data_line_names
```

```r
data <- na.omit(data)
```

```r
# TODO : English comments could be added.
# Usuwanie mirna o niskiej sumie zliczen. dla CCA (0,0,0,1) wyglada tak samo jak (0,0,0,100000), a to pierwsze moze byc wynikiem szumu
mirna_sum_threshold <- 20
mirna_sum_thr_mask <- rowSums(data) > mirna_sum_threshold
# TODO : This mask were not applied in oryginal script? Big reduction of columns.
data <- data[mirna_sum_thr_mask,]
```

```r
# 2log fold change
mirna_log2_treshold <- 4
# TODO : Check if lines'es names are OK!
l1 <- abs(log2((data[,"WM115"]+1)/(data[,"WM266-4"]+1)))
l2 <- abs(log2((data[,"WM793"]+1)/(data[,"WM1205Lu"]+1)))
mirna_log2_thr_mask <- ((l1 > mirna_log2_treshold) + (l2 > mirna_log2_treshold)) >= 1
data <- data[mirna_log2_thr_mask,]
```

```r
write.csv(data, "../data/clean_data/miRNA_count_data.csv", row.names = TRUE)
```
