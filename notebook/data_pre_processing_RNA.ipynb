```r
# TODO : From where those data are taken? 
raw_data <- read.csv("../data/raw_data/rna_total_cnt.csv", header=TRUE, row.names=NULL)
```

```r
raw_data <- data.frame(raw_data[!duplicated(raw_data[["X"]]), c(2:5)], row.names = raw_data[!duplicated(raw_data[["X"]]), "X"])
```

```r
raw_data <- na.omit(raw_data)
```

```r
# TODO : English comments could be added.
# Usuwanie mirna o niskiej sumie zliczen. dla CCA (0,0,0,1) wyglada tak samo jak (0,0,0,100000), a to pierwsze moze byc wynikiem szumu
rna_sum_threshold <- 20
rna_sum_thr_mask <- rowSums(raw_data) > rna_sum_threshold
# TODO : This mask were not applied in oryginal script? Not to big impact if comapred with 2log fold change filter.
raw_data <- raw_data[rna_sum_thr_mask,]
```

```r
# 2log fold change
rna_log2_treshold <- 4
# TODO : Change line. New data for processing. 
l1 <- abs(log2((raw_data[,"WM115"]+1)/(raw_data[,"WM266.4"]+1)))
l2 <- abs(log2((raw_data[,"WM793"]+1)/(raw_data[,"WM1205Lu"]+1)))
rna_log2_thr_mask <- ((l1 > rna_log2_treshold) + (l2 > rna_log2_treshold)) >= 1
data <- raw_data[rna_log2_thr_mask,]
```

```r
write.csv(data, "../data/clean_data/RAN_total_count_data.csv", row.names = TRUE)
```
